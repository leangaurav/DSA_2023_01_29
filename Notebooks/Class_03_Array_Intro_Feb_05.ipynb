    if(n == 0)

    int[] result = new int[2];
    int min = arr[0];
    int max = arr[0];

        if(arr[i] < min)
            min = arr[i];
        if(arr[i] > max)
            max = arr[i];

````python

def reverse(arr):
    i = 0
    j = len(arr) - 1
    while i < j:
        arr[i],arr[j] = arr[j],arr[i]
        i+=1
        j-=1

a = [1,2,3,4,5]
a.reverse()
print(a) 
reverse(a)
print(a) 
reverse(a)
print(a)

# TC: O(N)
````
