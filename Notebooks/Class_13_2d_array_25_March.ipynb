**C++**  
```C++
class Solution {
public:
    vector<vector<int>> transpose(vector<vector<int>>& matrix) {
        vector<vector<int>> ans;
        for(int i=0;i<matrix[0].size();i++){ // iterate column wise
            vector<int> v;
            for(int j=0;j<matrix.size();j++){ // row
                v.push_back(matrix[j][i]);
            }
            ans.push_back(v);
        }
        return ans;
    }
};
```

```Python
class Solution:
    def transpose(self, matrix: List[List[int]]) -> List[List[int]]:
        rows = len(matrix)
        cols = len(matrix[0])
        nm = [[0] * rows for _ in range(cols)] # pre-allocate

        for i in range(rows):
            for j in range(cols):
                nm[j][i] = matrix[i][j]
        return nm
```

**Java**  
```Java
class Solution {
    public int[][] transpose(int[][] matrix) {
        int a = matrix[0].length;
        int b = matrix.length;
        
        int[][] res = new int[a][b];
        for (int i=0; i < a; i++){
            for(int j = 0; j < b; j++){
                res[i][j] = matrix[j][i];
            }
        }
        return res;
    }
}
```

``` c++
class Solution {
public:
    void transpose(vector<vector<int>>& matrix) {
        for(int i=0;i<matrix.size();i++){ 

            for(int j=0;j<i;j++){
                int temp = matrix[j][i]; 
                matrix[j][i] = matrix[i][j];
                matrix[i][j] = temp;
            }
        }
        return matrix;
    }
};
```
